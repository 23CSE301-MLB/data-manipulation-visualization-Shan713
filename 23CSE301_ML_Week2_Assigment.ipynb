```python
import pandas as pd
df = pd.read_csv('Datasets/iris.csv')  # Reading the CSV file into a DataFrame
```

```python
df.head(10) # Displaying the first 10 rows of the DataFrame
```

```python
df.tail(5) # Displaying the last 5 rows of the DataFrame
```

```python
df.describe() # Displaying summary statistics of the DataFrame
```

```python
df.dtypes # Displaying the data types of each column
```

```python
df[df['SepalLengthCm']>5.0]  #Filtering rows where SepalLengthCm is greater than 5.0
```

```python
df['SepalTotalCm'] = df['SepalLengthCm'] + df['SepalWidthCm']  # Adding a new column 'SepalTotalCm' which is the sum of 'SepalLengthCm' and 'SepalWidthCm'
df
```

```python
df.groupby('Species')['SepalTotalCm'].mean() # Grouping by 'Species' and calculating the mean of a numerical column 'SepalTotalCm'
```

```python
df.sort_values(by='SepalLengthCm', ascending=False)  # Sorting the DataFrame by 'SepalLengthCm' in descending order
```

```python
df.sort_values(by=['SepalLengthCm','PetalLengthCm'], ascending=False)  # Sorting the DataFrame by 'SepalLengthCm' in descending order
```

```python
df.dropna(inplace=True)  # Checking for missing values in each column
```

```python
df.to_csv('Datasets/iris_modified.csv', index=False)  # Saving the modified DataFrame to a new CSV file
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")
```

```python
sns.kdeplot(data=df, x='PetalLengthCm', fill=True, color="blue")
plt.title('Distribution of Petal Length')
plt.show()
```

```python
sns.boxplot(data=df, x='Species', y='PetalLengthCm', hue='Species', palette='Set2')
plt.title('Boxplot of Petal Length by Species')
plt.show()
```

```python
sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species', style='Species', palette='Set1')
plt.title('Sepal Length vs Sepal Width by Species')
plt.show()
```

```python
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df[['Species','SepalLengthCm','SepalWidthCm']], hue='Species', palette='Set1')
plt.show()
```
